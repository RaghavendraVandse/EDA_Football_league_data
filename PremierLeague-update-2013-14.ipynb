```python
Referees = leaguestandings2013_2014.groupby('Referee')['Referee'].count()
Referees = Referees.sort_values(ascending = False)
Referees = pd.DataFrame(Referees)
Referees.rename(columns={'Referee':'Matches'})
```

```python
Referees.plot.bar()
```

```python
Refereesshownyellowathome = leaguestandings2013_2014.groupby('Referee')['HY'].sum()
Refereesshownyellowathome = Refereesshownyellowathome.sort_values(ascending = False)
Refereesshownyellowathome = pd.DataFrame(Refereesshownyellowathome)
Refereesshownyellowaway = leaguestandings2013_2014.groupby('Referee')['AY'].sum()
Refereesshownyellowaway = Refereesshownyellowaway.sort_values(ascending = False)
Refereesshownyellowaway = pd.DataFrame(Refereesshownyellowaway)
Refereesshownredathome = leaguestandings2013_2014.groupby('Referee')['HR'].sum()
Refereesshownredathome = Refereesshownredathome.sort_values(ascending = False)
Refereesshownredathome = pd.DataFrame(Refereesshownredathome)
Refereesshownredaway = leaguestandings2013_2014.groupby('Referee')['AR'].sum()
Refereesshownredaway = Refereesshownredaway.sort_values(ascending = False)
Refereesshownredaway = pd.DataFrame(Refereesshownredaway)

#Referees.rename(columns={'Referee':'Matches'})
```

```python
#Referees.rename(columns={'Referee':'Matches'})
Refereestandings = pd.concat([Referees,Refereesshownredaway,Refereesshownredathome,Refereesshownyellowaway,Refereesshownyellowathome],axis=1,sort=True)
Refereestandings['Yellowcards_shown'] = Refereestandings.AY+Refereestandings.HY
Refereestandings['Redcards_shown'] = Refereestandings.AR+Refereestandings.HR
Refereestandings.sort_values(by='Redcards_shown',ascending=False)
ToughestReferee=Refereestandings.rename(columns={'Referee':'Matches_officiated','AR':'Red Card shown to away player','HR':'Red Card shown to home player','AY':'Yellow Card shown to away player','HY':'Yellow Card shown to home player'})
ToughestReferee.sort_values(by='Redcards_shown',ascending=False)
#.sort_values(['Points'], ascending=[False])
```
